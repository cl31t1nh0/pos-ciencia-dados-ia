```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def plot_ds(ax, wls, xdata, ydata):
    wls_len = len(wls)
    for idx, rdata in xdata.iterrows():
        oct_value = ydata[idx]
        vals = rdata
        oct_array = np.full(wls_len, oct_value)

        ax.plot(oct_array, wls, vals)

path = '/Users/cleitonfraga/Downloads/dataset octangem.xlsx'
oct_df = pd.read_excel(path)

# Create the xdata y ydata (list of wavelengths and octane numbers)
oct_wls = np.array([ int(i) for i in oct_df.columns.values[:-1]])
oct_ydata = oct_df['Oct']
oct_xdata = oct_df[oct_df.columns.values[:-1]]

# Create a figure with two axes for 3D plotting
f,(ax) = plt.subplots(1, subplot_kw=dict(projection='3d'))

plot_ds(ax, oct_wls, oct_xdata, oct_ydata)
ax.set_title("Octane Dataset")
ax.set_xlabel("Octane Number")
ax.set_ylabel("Wavelength")
ax.set_zlabel("Value")

plt.show()
```
